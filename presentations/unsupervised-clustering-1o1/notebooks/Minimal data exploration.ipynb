```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import numpy as np
```

# load example

```python
digits = datasets.load_digits(return_X_y=True)
```

```python
matrix = digits[0]
```

# What is in there?

```python
for i in range(18):
    plt.subplot(3, 6, i+1)
    plt.imshow(matrix[i,:].reshape(8, 8), cmap='Greys')
    plt.grid()
    plt.xticks([i + 0.5 for i in range(8)], ['']*8)
    plt.yticks([i + 0.5 for i in range(8)], ['']*8)
```

## Distribution of values

```python
plt.figure(figsize=(12, 12))
for i in range(matrix.shape[1]):
    plt.subplot(8, 8, i + 1)
    plt.hist(matrix[:,i], bins=20, color='k', alpha=0.5)
plt.tight_layout()
```

## General picture

```python
    plt.imshow(np.sum(matrix, axis=0).reshape(8, 8), cmap='Greys')
    plt.colorbar()
```

## Normalization

```python
norm_matrix = StandardScaler().fit_transform(matrix)
```

```python
plt.imshow(np.sum(norm_matrix, axis=0).reshape(8, 8), cmap='Greys')
plt.colorbar()
```

## Data exploration (bis)

### before normalization

```python
X_plot = TSNE().fit_transform(matrix)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(*X_plot.T, 'k.', alpha=0.5)
```

### After normalization

```python
norm_X_plot = TSNE().fit_transform(norm_matrix)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(*norm_X_plot.T, 'k.', alpha=0.5)
```
